```python
################################################Importación de librerías
import pandas
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import figure
import datetime
################################################Creación de data Frame
Lluvia_DB = [
    '201801010000Lluv.csv',
    '201802010000Lluv.csv',
    '201803010000Lluv.csv',
    '201804010000Lluv.csv',
    '201805010000Lluv.csv',
    '201806010000Lluv.csv',
    '201807010000Lluv.csv',
    '201808010000Lluv.csv',
    '201809010000Lluv.csv',
    '201810010000Lluv.csv',
    '201811010000Lluv.csv',
    '201812010000Lluv.csv',
    '201901010000Lluv.csv',
    '201902010000Lluv.csv',
    '201903010000Lluv.csv',
    '201904010000Lluv.csv',
    '201905010000Lluv.csv',
    '201906010000Lluv.csv',
    '201907010000Lluv.csv',
    '201908010000Lluv.csv',
    '201909010000Lluv.csv',
    '201910010000Lluv.csv',
    '201911010000Lluv.csv',
    '201912010000Lluv.csv',
    '202001010000Lluv.csv',
    '202002010000Lluv.csv',
    '202003010000Lluv.csv',
    '202004010000Lluv.csv',
    '202005010000Lluv.csv',
    '202006010000Lluv.csv',
    '202007010000Lluv.csv',
    '202008010000Lluv.csv',
    '202009010000Lluv.csv',
    '202010010000Lluv.csv',
    '202011010000Lluv.csv',
    '202012010000Lluv.csv',
    '202101010000Lluv.csv',
    '202102010000Lluv.csv',
    '202103010000Lluv.csv',
    '202104010000Lluv.csv',
    '202105010000Lluv.csv',
    '202106010000Lluv.csv',
    '202107010000Lluv.csv',
    '202108010000Lluv.csv',
    '202109010000Lluv.csv',
    '202110010000Lluv.csv',
    '202111010000Lluv.csv',
    '202112010000Lluv.csv',
    '202201010000Lluv.csv',
    '202202010000Lluv.csv',
    '202203010000Lluv.csv',
    '202204010000Lluv.csv',
    '202205010000Lluv.csv',
    '202206010000Lluv.csv',
    '202207010000Lluv.csv',
    '202208010000Lluv.csv',
    '202209010000Lluv.csv'
]

def unpivot(file):
    df = pandas.read_csv(file, index_col=0, encoding='latin-1').reset_index()
    df = df.melt(id_vars=['LON','LAT','EDO','CLAVE','ESTACION'], var_name='MES', value_name='PRECIPITACION')
    return df

def append(Lluvia_DB):
    df = pandas.DataFrame()
    for db in Lluvia_DB:
        df = pandas.concat([df, unpivot(db)])
    df.dropna(subset=['PRECIPITACION'], inplace=True)
    return df

append(Lluvia_DB).to_csv('lluvias.csv', encoding="utf-8", index = False)
dataFrame = append(Lluvia_DB)
dataFrame
```

```python
############################################################Cambio de formato horario (YYYY-MM-DD)
dataFrame['MES'] = pandas.to_datetime(dataFrame['MES'])
dataFrame['MES2'] = dataFrame['MES'].dt.strftime('%Y-%d-%m')
dataFrame
```

```python
############################################################Limpieza de data frame (enfocado en el estado de Nuevo León)
porFechaNL=dataFrame[dataFrame.EDO=="NL"]
porFechaNL=porFechaNL[porFechaNL.PRECIPITACION.notnull()]
porFechaNL
```

```python
#############################################################Agrupación del promedio de Precipitacion por mes, por año en el estado de Nuevo León
NLmes19=porFechaNL.groupby(["MES2"])[["PRECIPITACION"]].mean().sort_values(by=["MES2"], ascending=True)
NLmes19
```

```python
########################################################### Linea de tiempo de precipitación 2019-2022 en el estado de Nuevo León
NLmes19.plot(kind="line", ylabel="mm", xlabel="2019-2022", title="Comparativa de Precipitación por Mes en Nuevo León (2019)", figsize=(30,8), fontsize=12, rot=1, grid=True)
```

```python
########################################################Separación por año de la precipitación  en el estado de Nuevo León
por2019=porFechaNL[porFechaNL.MES2.str.contains("2019")]
por2020=porFechaNL[porFechaNL.MES2.str.contains("2020")]
por2021=porFechaNL[porFechaNL.MES2.str.contains("2021")]
por2022=porFechaNL[porFechaNL.MES2.str.contains("2022")]
```

```python
group2019=por2019.groupby(["MES2"])[["PRECIPITACION"]].max()
group2020=por2020.groupby(["MES2"])[["PRECIPITACION"]].max()
group2021=por2021.groupby(["MES2"])[["PRECIPITACION"]].max()
group2022=por2022.groupby(["MES2"])[["PRECIPITACION"]].max()
group=["Enero","Febrero","Marzo","Abril","Mayo","Junio","Julio","Agosto","Septiembre","Octubre","Noviembre","Diciembre",]
```

```python
###################################################Gráfica comparativa anual (2019-2022) de la precipitación en el estado de nuevo León
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(15)
plt.plot(group,group2019.PRECIPITACION, label='2019')
plt.plot(group,group2020.PRECIPITACION,label='2020')
plt.plot(group,group2021.PRECIPITACION,label='2021')
plt.xlabel('MES DEL AÑO') 
plt.ylabel('PRECIPITACIÓN [mm]')
plt.title('Comparativa Anual de Precipitación en el Estado de Nuevo León')
plt.legend() 
plt.grid()
plt.show()
```
